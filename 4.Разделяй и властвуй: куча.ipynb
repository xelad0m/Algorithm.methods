### <a id='toc1_2_1_'></a>[Теоретическая задача для самостоятельной проверки: частичная сортировка](#toc0_)

Постройте алгоритм, который по данному массиву A[1…n] выводит его минимальные √n элементов в порядке возрастания (другими словами, выводит A′[1…√n]) за время O(n).

=> Построить мин кучу. В цикле от 1 до n\^1/2 доставать минимальный элемент. T(n) = O(n) + O(n\^1/2*log(n\^1/2))= O(n)


### <a id='toc1_2_2_'></a>[Теоретическая задача повышенной сложности для самостоятельной проверки: попарные суммы в возрастающем порядке](#toc0_)

Даны массивы A[1…n] B[1…n]. Мы хотим вывести все n\^2 сумм вида A[i]+B[j] в возрастающем порядке. Наивный способ — создать массив, содержащий все такие суммы, и отсортировать его. Соответствующий алгоритм имеет время работы O(n2log⁡n) и использует O(n2) памяти. Приведите алгоритм с таким же временем работы, который использует линейную память.

=> 
Alexey Udovenko

Сортируем A и B. Для каждого a∈A будем хранить позицию ia в B такую, что все пары (a,B[i]),i≥ia мы ещё не выдали. (Очевидно, что пары с элементом a мы будем выдавать в порядке увеличения элемента из B). Храним это всё в мин-куче, где ключ это a+B[ia]. Т.е. в куче можно хранить кортежи вида (a+B[ia],a,ia). Тогда вытаскивая элемент (a+B[ia],a,ia), выводим a+B[ia], увеличиваем ia до ia+1, и если он не вышел за рамки массива B, кладем в кучу (a+B[ia+1],a,ia+1).




## <a id='toc1_3_'></a>[Стабильная сортировка подсчётом](#toc0_)

Сортировка небольших по величине целых чисел

	Процедура CountSort(A[1. . .n])  
	{массивAсодержит целые числа от 1 до M}  
	создать массив B[1..M]←[0,0,..,0]  
	для j от 1 до n:  
		B[A[j]]←B[A[j]] + 1  
	для i от 2 до M: 
		B[i]←B[i] +B[i−1]  # количество элементов <= i
	для j от n до 1:  
		A′[B[A[j]]]←A[j]  
		B[A[j]]←B[A[j]]−1  

Время работы O(n+M)
Стабильность состоит в том, что последний цикл если идет в обратном порядке, то пусть и неразличимые равные элементы оказываются в том же порядке, что и были в исходном массиве.

## <a id='toc1_4_'></a>[Цифровая сортировка](#toc0_)
d раз вызывается сортировка подсчетом по каждому разряду начиная с младшего.  
И тут как раз нужна стабильность, т.к. отсортировав по младшему разряду, на следующем шаге числа станятся уже неразличимы для алгоритма и их порядок не должен измениться. Например: в массиве есть ..215..213..216.. => по первому разряду они станут ..213..215..216.. и последующие сортировки по следующим разрядом уже не должны изменить их взаимного расположения.

Время работы: O(nd), где d— число разрядов во входных числах.


### <a id='toc1_4_1_'></a>[Задача на программирование: сортировка подсчётом](#toc0_)

Первая строка содержит число 1≤n≤10\^4, вторая — n натуральных чисел, не превышающих 10. Выведите упорядоченную по неубыванию последовательность этих чисел.

```python
import random

INPUTS = """5
2 3 9 2 9"""

M = 10

def sort(A):
    n = len(A)
    B = [0] * (M + 1)
    A1 = [0] * n

    j = 0
    while j < n:
        try:
            B[A[j]] += 1
        except:
            print(j, len(A), A)
        j += 1

    i = 2
    while i <= M:
        B[i] = B[i] + B[i - 1]
        i += 1

    j = n - 1
    while j >= 0:
        A1[B[A[j]] - 1] = A[j]
        B[A[j]] -= 1
        j -= 1

    return A1

def test(n=10000):
    for l in range(1, n, n // 100):
        L = [random.randint(1, M) for _ in range(l)]
        try:
            assert sort(L) == sorted(L)
        except:
            print(L)


reader = (line for line in INPUTS.split('\n'))
n = int(next(reader))
A = list(map(int, next(reader).split()))

print(" ".join(list(map(str, sort(A)))))

test()
```

### <a id='toc1_4_2_'></a>[Теоретическая задача повышенной сложности для самостоятельной проверки: локальный минимум за сублинейное время](#toc0_)

Локальным минимумом матрицы называется элемент, значение которого не превосходит значений всех его соседей (считаем, что у внутреннего элемента матрицы четыре соседа, у элемента на границе — три, а у углового элемента — два). Постройте алгоритм, который по данной матрице размера n×n, содержащей попарно различные натуральные числа, находит (любой) локальный минимум за время O(n) (такое время называется сублинейным, поскольку оно по порядку меньше размера входа).

=> Попарно различны (все разные), значит есть минимальный эл-т, значит есть по крайней мере глобальный минимум. Также, то что все числа различны, переходя к минимальному соседу случайно выбранного начального числа мы будет гарантированного приближаться к локальному минимуму. Выбирая минимальный из 4-х соседей мы минимум в 4 раза сокращаем возможный путь к локальному минимуму. Максимальный путь 2*n шагов, т.е. O(n).

```python
# что видно даже невооруженным глазом:
import random

n = 5
E = list(range(0, n * n))
random.shuffle(E)
M = [E[i:i + n] for i in range(0, n * n, n)]

M
```
